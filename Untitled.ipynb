```python
pieces=[]
for year in range(1880,2022):
    path=f"dataset/babynames/yob{year}.txt"
    data=pd.read_csv(path,names=["name","sex","births"]) 
    data["year"]=year
    pieces.append(data)
names=pd.concat(pieces,ignore_index=True)
```

```python
names
```

```python
total_birth=names.pivot_table(values="births",index="year",columns="sex",aggfunc=sum)
```

```python
total_birth.plot(title="Total births by sex and year")
```

```python
names = names.groupby(['year', 'sex']).apply(
    lambda group: group.assign(proportion=group['births'] / group['births'].sum())
).reset_index(drop=True)
```

```python
names
```

```python
names.groupby(["year","sex"])["proportion"].sum() # the proportions add up to 1, 
                                                  # indicating the proportions were calculated correctly
```

```python
grouped = names.groupby(['year', 'sex'])
Top1000_name = grouped.apply(
    lambda group: group.sort_values("births", ascending=False).head(1000)
).reset_index(drop=True)
```

```python
Top1000_name=Top1000_name.reset_index(drop=True)
```

```python
boys=Top1000_name[Top1000_name["sex"]=="M"] 
girls=Top1000_name[Top1000_name["sex"]=="F"]
```

```python
# calculates the number of top names (sorted by proportion) needed to reach at least 50%
diversity = (Top1000_name.sort_values(by="proportion", ascending=False)
                     .groupby(["year", "sex"])["proportion"]
                     .apply(lambda x: x.cumsum().searchsorted(0.5) + 1)
                     .unstack())
```

```python
diversity.head()
```

```python
diversity.plot(title="Number of population name in 50%")
```
